```python
final.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
#convert to a dataframe
final.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
```

```python
#renaming 0 column
ratings = final.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,4)

sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,4)

#here hue and pallete parameter is used to make sure rating and color is consistent with the dataset
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['black','red','orange','yellow','green','green'])
```

```python
#count plot

#no. of records
sns.countplot(x='Rating color', data = ratings, palette = ['black','red','orange','yellow','green','green'])
```
